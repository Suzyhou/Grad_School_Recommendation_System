# T-test

<h3>T test on GPA</h3>
<h4>We would like to compare the difference between student's who got admitted to student's who got rejected. since we are having limited features, we will use T-test to compare this two groups</h4>

```python
df_ttest = df[df['gpa'].notna()]
```

```python
accepted = df_ttest.loc[(df_ttest['decision'] == 'Accepted') & (df_ttest['major'] == 'Computer Science'),'gpa'].to_numpy()
accepted
```

```python
rejected = df_ttest.loc[(df_ttest['decision'] == 'Rejected')& (df_ttest['major'] == 'Computer Science'),'gpa'].to_numpy()
rejected
```

```python
stats.ttest_ind(a=accepted,b=rejected,equal_var=True)
```

<h4>The p value obtained from the t-test is significant (p < 0.05), we reject the null hypothesis that there's no difference between the mans. And therefore, we conclude that the gpa of students who got accpeted is significantly different than students who got rejected.</h4>

<h3>T test on GRE</h3>


```python
df_ttest2 = df[df['gre'].notna()]
```

```python
accepted2 = df_ttest2.loc[(df_ttest2['decision'] == 'Accepted') & (df_ttest2['major'] == 'Computer Science'),'gre'].to_numpy()
accepted2
```

```python
rejected2 = df_ttest2.loc[(df_ttest2['decision'] == 'Rejected')& (df_ttest2['major'] == 'Computer Science'),'gre'].to_numpy()
rejected2
```

```python
stats.ttest_ind(a=accepted2,b=rejected2,equal_var=True)
```

<h4> Here p = 0.26 which is larger than 0.05, so we failed to reject the hypothesis that there's no difference between the GRE means between the two groups. And we conclude that there is no enough evidence that students who got admitted have significant difference between students who got accepted.

```python
def get_relevantData(GREcheck,majorCheck):
    df1 = df.loc[df['decision'] == 'Accepted']
    if majorCheck != '':
        df1_accepted = df1.loc[df1['major'] == majorCheck]
    if GREcheck == True:
        df1_accepted = df1_accepted[~(df1_accepted['gre'].isnull())]
        df1_accepted = df1_accepted[~(df1_accepted['gpa'].isnull())]
        df1_accepted = df1_accepted[['gpa','gre','research','school']]
    else:
        df1_accepted = df1_accepted[~(df1_accepted['gpa'].isnull())]
        df1_accepted = df1_accepted[['gpa','research','school']]
    #df1_accepted = df1_accepted.reset_index().drop(["index"], axis=1)
    
    print(df1_accepted.shape)
    return df1_accepted
```

# Admission Recommender System with KNN

```python
def euclideanDistance(data1, data2, length):
    distance = 0
    for x in range(length):
        distance += np.square(data1[x] - data2[x])
    return np.sqrt(distance)
```

```python
def admissionRecommender(admissionData,inputData, topk):
    distances = []
    length = inputData.shape[1]

    for idx in range(len(admissionData)):
        distance = euclideanDistance(inputData, admissionData.iloc[idx], length)
        distances.append(distance[0])
    admissionData['distance'] = distances
    top_list = (admissionData.sort_values("distance", ascending=True)
                [["school","distance"]]
                .drop_duplicates()[:topk])
    return top_list
```

```python
testSet = [[4, 170, 0]]
test = pd.DataFrame(testSet)
test.shape
```

```python
recommendSchool = admissionRecommender(accepted, test, 5)
```
